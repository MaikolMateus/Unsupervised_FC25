```python
# Publicamos los datos en la web

url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vSg9jig5EsHmpD-cE0Rvzst6k84mCdiYfcyEFw9xesoomi_jQ0W1_fF2nfSnFlZonFirLpu-mpNkxT3/pub?output=csv'

data = pd.read_csv(url)
data.head()
```

```python
# Reorganizamos las bases de datos

columnas_objetivo = ['Name', 'OVR', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY', 'Acceleration',
       'Sprint Speed', 'Positioning', 'Finishing', 'Shot Power', 'Long Shots',
       'Volleys', 'Penalties', 'Vision', 'Crossing', 'Free Kick Accuracy',
       'Short Passing', 'Long Passing', 'Curve', 'Dribbling', 'Agility',
       'Balance', 'Reactions', 'Ball Control', 'Composure', 'Interceptions',
       'Heading Accuracy', 'Def Awareness', 'Standing Tackle',
       'Sliding Tackle', 'Jumping', 'Stamina', 'Strength', 'Aggression',
       'Weak foot', 'Skill moves', 'Age', 'GK Diving', 'GK Handling',
       'GK Kicking', 'GK Positioning', 'GK Reflexes', 'Preferred foot',]

columnas = data.columns

for col in columnas:
  if col not in columnas_objetivo:
    data.drop(col,axis = 1,inplace = True)
  else:
    data[col] = data[col]

data.head()
```

```python
# Verificar datos faltantes
data.isna().sum()
```

```python
#Imputar datos faltantes para las categorías de GK cambiando NaN por 0

gk_columns = ['GK Diving', 'GK Handling', 'GK Kicking', 'GK Positioning', 'GK Reflexes']
data[gk_columns] = data[gk_columns].fillna(0)
```

```python
# Extraemos los datos

X = data.iloc[:,1:].values
```

```python
# Codificar el pie preferido como variable numérica
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

X[:, -7] = le.fit_transform(X[:, -7])

# Verificamos
X[:, -7]
```

```python
# Data Frame a valores numéricos
df = pd.DataFrame(X)
for col in df.columns:
    # Convertir a numérico, forzando errores a NaN
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
X = df.values
X
```

```python
# Empezamos a definir las librerías para el método con sklearn
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
imputer.fit(X)
X = imputer.transform(X)
X
```

# Montaje de Algoritmo de Clasificación K-means

##PCA

```python
# Definimos las librerías
from sklearn.decomposition import PCA

# Crear un objeto PCA y ajustar los datos de las tres principales componentes
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)
```

```python
# Verificamos la varianza de los datos
varianza = pca.explained_variance_ratio_
print(sum(varianza))
```

```python
# método del codo
# método del codo
from sklearn.cluster import KMeans
## Inicializo donde almaceno el WCCS(k)
WCSS = []

for k in range(1,11):
  kpp = KMeans(n_clusters= k,
               #init = 'k-means++',
               random_state=42, n_init = 'auto')
  kpp.fit(X)
  WCSS.append(kpp.inertia_)

plt.figure()
plt.plot(range(1,11), WCSS, 'o--', color = 'red')
plt.title('Método del Codo')
plt.xlabel('k')
plt.ylabel('WCSS')
plt.grid(True)
plt.grid(linestyle = '-')
plt.show()

# Según la interpretación de los nodos, encontramos el valor de cluster optimo en x=4
```

## Usando la librería Sklearn

```python
# Definimos la librería
from sklearn.cluster import KMeans

# Definimos el algoritmo
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(X) # Organizar y asociando los datos en 4 grupos.
```

```python
# Definimos los centroides
centroides_kmeans = kmeans.cluster_centers_
centroides_kmeans = pca.transform(centroides_kmeans)
centroides_kmeans
```

```python
# Definimos la función de gráfica de clusters y centroides 2D

def graficar_clusters_PCA2D(clusters, metodo, centroids):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1],
                    hue=clusters, palette='deep',
                    alpha=0.6, s = 100)
    plt.scatter(centroids[:,0], centroids[:,1], c= 'black', s = 100, label = 'centroides')
    plt.title(f'Visualización 2D de la Clusterización (PCA) {metodo}', fontsize=16)
    plt.xlabel('Componente Principal 1')
    plt.ylabel('Componente Principal 2')
    plt.grid(True)
    plt.grid(linestyle = '-', alpha  = 0.5)
    plt.legend(title='Cluster')
    plt.show()
```

```python
# Definimos la función de gráfica de clusters y centroides 3D
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import seaborn as sns

def graficar_clusters_PCA3D(cluster,metodo):
    #Crear el grafico 3D
    fig = plt.figure(figsize=(15, 8))
    ax = fig.add_subplot(111, projection='3d')

   #Graficar los puntos con borde blanco
    scatter = ax.scatter(X_pca[:, 0],X_pca[:, 1],X_pca[:, 2],
                         c=cluster,cmap='viridis', s=100,
                        edgecolor ='white',alpha = 0.6)
    #Etiquetas de los ejes y titulo
    ax.set_title(f'Visualización 3D de la Clusterización (PCA) (metodo)', fontsize=16)
    ax.set_xlabel('Componente Principal 1')
    ax.set_ylabel('Componente Principal 2')
    ax.set_zlabel('Componente Principal 3')

    #Añadir leyenda
    legend1 = ax.legend(*scatter.legend_elements(), title = "Clusters")
    ax.add_artist(legend1)

    #Mostrar grafico
    plt.show()
```
